# Oscillation

- find fixed point
- test stability of the model

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Equation 1: $P_{n+1} = 2P_{n} - 3$

```python
# plot the graph 
a = 2
b = -3
n = np.arange(0, 11)
print("n: ", n)
for p0 in [-2, 3]:
  ps = []
  name = p0
  for i in range(len(n)):
    pn = a*p0 + b
    p0 = pn
    ps.append(pn)
  plt.plot(n, ps, label = "p0="+str(name))
  print("p_n (p_0="+str(name)+"): ", ps)
plt.title("Equation 1")
plt.xlabel("n")
plt.ylabel("p_n")
plt.legend()
plt.show()
```
